```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Datasets/N_de_mencoes.txt', parse_dates=(['Data']))
colunas = ['Semana de', 'N. de notícias', 'Sem menção', 'Com menção Dr. Sarrubbo']
df.columns = colunas
df
```

```python
import numpy as np

x = df['Sem menção'].sum()
y = df['Com menção Dr. Sarrubbo'].sum()
print(f'Notícias sem menção: {int(x)}\nNotícias com menção ao Dr. Sarrubbo: {int(y)}\n\n\n')

y = np.array([x, y])
etiquetas = ['Sem menção','Com menção Dr. Sarrubbo']

plt.pie(y, labels = etiquetas, explode=(0, 0.3), autopct='%1.1f%%')
plt.gca().update(dict(title='Notícias do portal - Gestão Dr. Sarrubbo - até 08/05/2022'))
plt.savefig('pizza_noticias.jpg')
plt.show()
```

```python
sem_mencao = df['N. de notícias'][df['Com menção Dr. Sarrubbo']==0].count()
total = df['N. de notícias'].count()
com_mencao = total - sem_mencao
print(f'Total de semanas: {total}\n')

x = sem_mencao
y = com_mencao
print(f'Semanas sem menção: {x}\nSemanas com menção ao Dr. Sarrubbo: {y}\n\n\n')

y = np.array([x, y])
etiquetas = ['Sem menção','Com menção ao Dr. Sarrubbo']

plt.pie(y, labels = etiquetas, explode=(0, 0.3), autopct='%1.1f%%')
plt.gca().update(dict(title='Notícias do portal - Gestão Dr. Sarrubbo - até 08/05/2022'))
plt.savefig('pizza_sem.jpg')
plt.show()
```

```python
df1 = df[['Sem menção',	'Com menção Dr. Sarrubbo']]
df1.plot.area(title=('Notícias em 109 semanas - até 08/05/2022'), figsize=(12, 4), subplots=True)
plt.savefig('area_109semanas.jpg')
plt.show()
```

```python
df1.plot(title='Menções em 109 semanas - i.e. até 08/05/2022', figsize=(12, 4), subplots=True)
plt.savefig('subplots_109semanas.jpg')
plt.show()
```
